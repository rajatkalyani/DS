```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import random
from pprint import pprint
```

```python
df = df.drop("Id", axis=1)
df = df.rename(columns={"species": "label"})
```

```python
# writing the train_test_split function:
def train_test_split(df, test_size):
    if isinstance(test_size, float):
        test_size = round(test_size * len(df))
    indices = df.index.tolist()
    test_indices = random.sample(population = indices, k = test_size)
    test_df = df.loc[test_indices]
    train_df = df.drop(test_indices)
    return train_df, test_df
```

```python
# we are using random.seed(0) to make sure every run the random number of test data frame should be same
# normally we have to select 20 flowers randomly out of 150. for that we have to use random variable
random.seed(0)
train_df, test_df = train_test_split(df, test_size=20)
```

```python
# Let's write the function for the data pure function
def check_purity(data):
    
    label_column = data[:, -1]
    unique_classes = np.unique(label_column)

    if len(unique_classes) == 1:
        return True
    else:
        return False
    
```

```python
# if we want to verify the function is giving proper output or not, then we can pass the train_df.values
check_purity(train_df.values)
# If we want to verify the data having petal_width < 0.8 
check_purity(train_df[train_df.petal_width < 0.8].values)
# If we want to verify the data having petal_width > 0.8 
check_purity(train_df[train_df.petal_width > 0.8].values)
display(check_purity(train_df.values),check_purity(train_df[train_df.petal_width < 0.8].values),check_purity(train_df[train_df.petal_width > 0.8].values))
```

```python
# If we want to see check the uniqueness of the values preset in the label column and also want the count.
# Then we can use the np.unique function with additional argument return_counts = True
label_column = data[:, -1]
unique_classes, counts_unique_classes = np.unique(label_column,return_counts = True)
display(unique_classes, counts_unique_classes)
```

```python
# If we want to see the largest value within the numpy one dimensional aray counts_unique_classes
# we can use argmax(). The argmax() will return the index of the maximum value
index = counts_unique_classes.argmax()
index
# we can use the index inside the unique_classes to get the species which has the height value
classification = unique_classes[index]
display(classification,index)
```

```python
# let's write the classification function:
def classify_data(data):
    
    label_column = data[:, -1]
    unique_classes, counts_unique_classes = np.unique(label_column, return_counts=True)

    index = counts_unique_classes.argmax()
    classification = unique_classes[index]
    
    return classification
```

```python
# let's verify the function working properly or not.
op1 = classify_data(train_df[train_df.petal_width < 0.8].values)
# if we pass < 1.2 still it will show Iris-setosa as majority are Iris-setosa
op2 = classify_data(train_df[train_df.petal_width < 1.2].values)
# If we want to verify a length in between some range
op3 = classify_data(train_df[(train_df.petal_width > 0.8) & (train_df.petal_width < 2 )].values)
display(op1,op2,op3)
```

```python
# now we need to write Potential splits
def get_potential_splits(data):
    
    potential_splits = {}
    _, n_columns = data.shape
    for column_index in range(n_columns - 1):        # excluding the last column which is the label
        potential_splits[column_index] = []
        values = data[:, column_index]
        unique_values = np.unique(values)

        for index in range(len(unique_values)):
            if index != 0:
                current_value = unique_values[index]
                previous_value = unique_values[index - 1]
                potential_split = (current_value + previous_value) / 2
                
                potential_splits[column_index].append(potential_split)
    
    return potential_splits
```

```python
# let's verify if the function is working fine or not
potential_splits = get_potential_splits(train_df.values)
# It will give you the index of the column and the ptential split values
```

```python
# we can visualize the potential_splits with the scatter split
sns.lmplot(data = train_df,x = 'petal_width', y = 'petal_length',hue = 'label',fit_reg = False,size = 6,aspect=1.5)
# Now let's draw the potential_split on top of the seaborn scatter plot
plt.vlines(x = potential_splits[3],ymin =1,ymax = 7)
```

```python
# If we want to draw  potentail_lines for petal_length .
sns.lmplot(data = train_df,x = 'petal_width', y = 'petal_length',hue = 'label',fit_reg = False,size = 6,aspect=1.5)
# Now let's draw the potential_split on top of the seaborn scatter plot
plt.hlines(y = potential_splits[2],xmin =0,xmax = 2.5)
```
