```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men])
```

```python
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men],bins=[80,100,125,150])
```

```python
# As the bins are not seperated we have to seperate by using rwidth.
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men],bins=[80,100,125,150],rwidth=0.95)
```

```python
# similarly we can plot for women
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_women],bins=[80,100,125,150],rwidth=0.95)
```

```python
# Histogram are the bar plot in which the y-axis contains the frequency of a variable
# let's draw both the histogram together
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins =[80,100,125,150],rwidth=0.95)
plt.legend()
```

```python
# as it is difficult to figureout which bar is for men and which is for women, need to add label.
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins =[80,100,125,150],rwidth=0.95,label =['men','women'])
plt.legend()
```

```python
# we can change the color
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins =[80,100,125,150],rwidth=0.95,color = ['green','orange'],label =['men','women'])
plt.legend()
```

```python
# If we want to plot this is in horizontal,we have to add orientation
plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins =[80,100,125,150],rwidth=0.95,color = ['green','orange'],label =['men','women'],orientation='horizontal')
plt.legend()
```
