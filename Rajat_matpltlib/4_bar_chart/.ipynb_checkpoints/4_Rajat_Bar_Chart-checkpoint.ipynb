```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# if we want to plot a bar chart we can do directly plt,.bar( company,revenu)
# But the problem is the bar chart will not accept the string values like 'Google','AMZN'
# so we need to convert the strings to number, need to use xpostion.
# If the revenune would have in string then we should have created yposition
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
xpos = np.arange(len(company))
xpos
```

```python
plt.xlabel('companies')
plt.ylabel('revenue')
plt.title('companies ~ reveune')
plt.bar(xpos,revenue)
```

```python
# we we are not able to identity which x-value is for which compnay we have to use now x-tick function
plt.xticks(xpos,company)
plt.xlabel("company-names")
plt.ylabel("revenue-in-usd")
plt.title('US-TECH-STOCKS')
plt.bar(xpos,revenue,label='revenue')
plt.legend()
```

```python
# If we want to add the profit to the graph, then for simplicity we can add one more line of plt.bar
# but in this way, one will come on top of other as the x-position is same for both profit and revenue
plt.xticks(xpos,company)
plt.xlabel("company-names")
plt.ylabel("revenue-in-usd")
plt.title('US-TECH-STOCKS')
plt.bar(xpos,revenue,label='revenue')
plt.bar(xpos,profit,label='profit')
plt.legend()
```

```python
# as we know the xpos is a numpy array, we can use the addition or substraction element wise
plt.xticks(xpos,company)
plt.xlabel("company-names")
plt.ylabel("revenue-in-usd")
plt.title('US-TECH-STOCKS')
plt.bar(xpos-0.2,revenue,label='revenue')
plt.bar(xpos+0.2,profit,label='profit')
plt.legend()
```

```python
# Stil the bars are not seperated, so we can use width argument to do this
plt.xticks(xpos,company)
plt.xlabel("company-names")
plt.ylabel("revenue-in-usd")
plt.title('US-TECH-STOCKS')
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
# If we want to draw the bar chart in horizontal fasion
# here we have to remove the width and the x-label and y-label
# as it is horizontal, we have use yticks instead of x-ticks
plt.yticks(xpos,company)
plt.title('US-TECH-STOCKS')
plt.barh(xpos-0.2,revenue,label='revenue')
plt.barh(xpos+0.2,profit,label='profit')
plt.legend()
```
