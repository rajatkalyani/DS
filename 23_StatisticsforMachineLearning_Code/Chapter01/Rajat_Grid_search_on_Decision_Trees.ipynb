```python
# As in the normal python or panda code we donot have standard function to split data into 
# Train, Validation and Test data, we have to use the split function twice for this and for this we have to write
# customized function

def data_split(data,trf = 0.5, vlf = 0.25,tsf = 0.25):
    nrows = data.shape[0]
    trnr = int(nrows * trf )
    vlnr = int(nrows * vlf)
    
    tr_data,rmng = train_test_split(data,train_size = trnr,random_state=42)
    vl_data, ts_data = train_test_split(rmng,train_size = vlnr,random_state=45)
    return (tr_data,vl_data,ts_data)

    
    
```

```python
train_data, validation_data, test_data = data_split(originial_data,trf=0.5,vlf=0.25,tsf=0.25)
```
