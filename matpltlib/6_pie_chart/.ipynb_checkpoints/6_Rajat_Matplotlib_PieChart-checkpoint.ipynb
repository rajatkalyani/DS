```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
# let's plot a pie chart
# here we are not able to decide how much percentage is expense of total expense
plt.pie(exp_vals,labels =exp_labels)
```

```python
# sometimes the pie chart may not come as a proper cir cle, it may come as ellipse
# so we have to use axis
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels)
```

```python
# If we want to avoid the extra print as [<matplotlib.patches.Wedge at 0x1145e0898>
# then we can use plt.show()
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels)
plt.show()
```

```python
# As we able to see a small pie chart due to its small radius. If we want a bigger diagram then 
# need to increase the radius size
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius=1.5)
plt.show()
```

```python
# If we want the percentage to display, we can use the argument autopct
# The format of autopct is %f%%.
# If we want the percentage upto two decimal, the we can write %0.2f%%
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius=1.5,autopct = '%0.2f%%')
plt.show()
```

```python
# If we donot want float then make %0.0f%%
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius=1.5,autopct = '%0.0f%%')
plt.show()
```

```python
# If we want shadow to be displayed in the pie chart then include the argument shadow=True.
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius=1.5,autopct = '%0.0f%%',shadow=True)
plt.show()
```

```python
# If we want to take out the pie means each one is split from each other
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius=1.5,autopct = '%0.0f%%',shadow=True,explode = [0,0.1,0,0.2,0])
plt.show()
```

```python
# If we want to start the pie shart to start from a differnt angle, that homrent should in differnt angle
# we can use attribute startangle
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius = 1.5 ,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0]
        ,startangle = 45)
plt.show()
```

```python
plt.axis = ('equal')
plt.pie(exp_vals,labels =exp_labels,radius = 1.5 ,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0,0,0]
        ,startangle = 45)
plt.show()
```
