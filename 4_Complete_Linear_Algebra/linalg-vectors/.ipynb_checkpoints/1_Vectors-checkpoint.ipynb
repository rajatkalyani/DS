```python
# Vector-vector multiplication: the dot product
## many ways to compute the dot product

v1 = np.array([ 1,  2,  3,  4,  5,  6 ])
v2 = np.array([ 0, -4, -3,  6,  5,  9])
# method 1
dp = sum( np.multiply(v1,v2) )

# method 2
dp = np.dot( v1,v2 )

# method 3
dp = np.matmul( v1,v2 )

# method 4
dp = 0  # initialize

# loop over elements
for i in range(0,len(v1)):
    
    # multiply corresponding element and sum
    dp = dp + v1[i]*v2[i]
    
print(dp)

```

```python
#Vector length
# a vector
v1 = np.array([ 1, 2, 3, 4, 5, 6, ])

# methods 1-4, just like with the regular dot product, e.g.:
vl = np.sqrt( sum( np.multiply(v1,v1)) )

# method 5: take the norm
vl = np.linalg.norm(v1)

# method 6 : using direct squre function instead of np.multiply
v1 = np.sqrt( sum( np.square(v1)) )

print(vl)
```
