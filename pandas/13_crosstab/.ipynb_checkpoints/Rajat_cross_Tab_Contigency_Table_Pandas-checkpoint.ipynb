```python
import os
os.chdir('/Users/rajatpa/Documents/DS/code_basics/py/pandas/13_crosstab')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("survey.xls")
df
```

```python
# cross tab is a method avaiable in pandas not in DataFrame. so we have to use pd and need to pass the DataFrame
pd.crosstab(df.Nationality,df.Handedness,margins = True)
```

```python
pd.crosstab(df.Sex,df.Handedness,margins = True)
```

```python
# want to multiple things in columns
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins = True)
```

```python
# If we want to check the values in the percentage 
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins = True,normalize="index")
```

```python
# Aggfunc and Values
# If we want to know the average Age of the male and female of left and write handed
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
