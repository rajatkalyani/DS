```python
# but we need to convert the types to date type.
df1 = pd.read_csv('/Users/rajatpa/Documents/DS/code_basics/py/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv',parse_dates = ['day'])
df1
```

```python
# but the way we handle the NA value is may not proper. we can fill the previous month value to the NA
df2 = df1.fillna(method ='ffill')
df2
# we can also use bfill instead of ffill if we want to do back ward copy
# we can also use the axis = 'columns' or axis = 'rows' also
```

```python
# if we want to fill the NA values with some new values but if we ware using the ffill ,then we want to fill only one
# place though there are many NAN are there continously. we can use limit argument.
df3 = df1.fillna(method ='ffill',limit = 1)
df3
```

```python
# instead of forward fill we can use interpolation method
# basically it will use Linear Regression kind algorithm and will fill the values
df3 = df1.interpolate()
df3
```

```python
# we can use many arguments inside the interpolate() method. let's use time
df4 = df1.interpolate(method = 'time')
df4
```

```python
# If we want to drop NA without any further processing
df4 = df1.dropna()
display(df1,df4)
# It is droping the rows if it has even single NA.
```

```python
# if want to drop NA if the row has all NAs
df5 = df1.dropna(how = 'all')
display(df1,df5)
```

```python
# the above can be written using argument threshold
# Here it will not drop the row if it has at least one valid value.
df6 = df1.dropna(thresh = 1)
display(df1,df6)
```

```python
# the above can be written using argument threshold
# Here it will not drop the row if it has at least two valid value.
df7 = df1.dropna(thresh = 2)
df7
```

```python
# If we want to add or insert the missing data index.
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)
df8 = df1.reindex(idx)
display(df1,df8)
```
