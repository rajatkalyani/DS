```python
df.pivot(index='date',columns="city")
```

```python
# with the help of pivot table we can do the aggregation and average details and etc
df1 = pd.read_csv("weather2.csv")
df1
```

```python
df2 = df1.pivot_table(index="city",columns="date")
df2
```

```python
# If instead of average i want the sum then need to define the agreegate functions accordingly
df3 = df1.pivot_table(index="city",columns="date", aggfunc='sum',margins = True)
df3
```

```python
df4 = df1.pivot_table(index="city",columns="date", aggfunc='count',margins = True)
df4
```
