```python
import os
os.chdir('//Users/rajatpa/Documents/DS/code_basics/py/pandas/9_merge')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
# in the pd.concat the common columns are normally repated. in merge we can reduce this to some extent
df3 = pd.merge(df1,df2,on = 'city')
df3
```

```python
# what would have happened if we would have used concat instead of merge
dfc = pd.concat([df1,df2],axis=1)
dfc1 = pd.concat([df1,df2],axis=0)
display(dfc,dfc1)
```

```python
# For concatinating the Data Frame 
df1i = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
},index = [0,1,2])
df1i

df2i = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
},index = [0,1,2])
df2i

display(df1i,df2i)
```

```python
display(pd.concat([df1i,df2i],axis =0),pd.concat([df1i,df2i],axis =1))
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando",'baltimore'],
    "temperature": [21,14,35,32]
})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","San Franisco"],
    "humidity": [65,68,71],
})
df5
```

```python
df6 = pd.merge(df4,df5,on = 'city')
# df6 = pd.merge(df4,df5,on = 'city',how = 'inner')
df6
# By default the merge do a intersection. This is called as inner join.
```

```python
# If we want to do the outer join or the union 
# If we want to know from which data frame the elemnts are came then we can use indicator = True
df7 = pd.merge(df4,df5,on = 'city',how = 'outer',indicator = True)
display(df4,df5,df7)
# it shows that the humidity is NaN for orlando and orlando is a left data Frame
# San Franisco is right Data Frame and temperature is NaN for it
```

```python
# If we use left joint, then it will consider the 1st DataFrame and the intersection with Second DataFrame
df8 = pd.merge(df4,df5,on = 'city',how = 'left')
display(df4,df5,df8)
```

```python
# If we use right joint, then it will consider the 2nd DataFrame and the intersection with first DataFrame
df9 = pd.merge(df4,df5,on = 'city',how = 'right')
display(df4,df5,df9)
```

```python
# Use of Suffixes
df10 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df10
df11 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df11

display(df10,df11)
```

```python
# Here in both the Data Frame df10 and df11 all the column names are same or common. 
# if we do the normal merge, it will repeat

df12 = pd.merge(df10,df11,on = 'city',how = 'outer')
df16 = pd.merge(df10,df11,on = 'city',how = 'inner')
display(df10,df11,df12,df16)
```

```python
display(df12.index,df12.columns,df12.ix[[0,1,2],["city",'temperature_x']],df16.ix[[0,1,2],["city",'temperature_x']])
```

```python
display(df12.loc[[0,1,2],["city",'temperature_x']],df16.loc[[0,1,2],["city",'temperature_x']])
```

```python
# Instead of _x and _y if we want to use the suffix as left and right DataFrame
df13 = pd.merge(df10,df11,on = 'city',how = 'outer',suffixes = ('_left','_right') )
df13
```

```python
df14 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df14.set_index('city',inplace=True)
df14
df15 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df15.set_index('city',inplace=True)
df15

display(df14,df15)
```

```python
# not sure the use of lsuffix='_r', rsuffix='_l' here 
df17 = df14.join(df15)
df18 = df14.join(df15,lsuffix='_r', rsuffix='_l')
display(df17,df18)
```
