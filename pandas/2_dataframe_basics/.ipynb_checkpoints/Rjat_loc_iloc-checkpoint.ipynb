```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head(10)
```

```python
# .loc is for filtering row and selecting columns by label.
# label means row by its index and column by column names.
# So .loc is for selecting things by label
# the format is .loc[what rows i want, what columns i want]
# If i want only row Zero and all columns then the syntax is .loc[0,:]
display(ufo.loc[0,:],ufo.loc[[0],:])
```

```python
# If want some specific rows 
# incase of .loc if we use : for the range everything is included means the start and end. for eaxmple 0:3
# means it starts at 0 and ends at 3 not at 2 as incase of normal python
# incase of range we canot use the double bracket for Data Frame : ufo.loc[0:2,:])
display(ufo.loc[[0,1,2],:],ufo.loc[0:2,:])
```

```python
# if want zeroth,1st and 2nd row and only city columns
display(ufo.loc[[0,1,2],["City"]],ufo.loc[[0,1,2],"City"])
```

```python
# If i want all rows and selected city
display(ufo.loc[:,['City','Shape Reported']])
```

```python
# here we want all rows but few columns but using the range operator. the range should be continous
display(ufo.loc[:,'Shape Reported':'Time'])
```

```python
# if want to use the range for row and columns
display(ufo.loc[0:3,'City':'State'])
```

```python
# the above can be written without using the .loc mentod 
ufo.head(3).drop('Time',axis=1)
```

```python
# If i want all the rows which have only city with Oakland
ufo[ufo.City == 'Oakland']
```

```python
# If i want to do the same i,e want all the rows which have only city with Oakland using .loc
ufo.loc[ufo.City == 'Oakland',:]
```

```python
# let's say we want all the rows having Oakland(it will give row index) but having coulmn State.
display(ufo.loc[ufo.City == 'Oakland',"State"],ufo[ufo.City =='Oakland'].State)
```

```python
# the .iloc is filtering rows and and selecting columns by integer postions not by Label which .loc use to do.
display(ufo.iloc[:,[0,3]])
```

```python
# If we want rows from 0 to 4 and coulmn 0 to 2 
# Here the range works like the normal python range so if we give 0:4, then 0 is included but 4 is excluded.
display(ufo.head(3),ufo.iloc[0:5,0:3])
```

```python
display(ufo.columns,ufo.head(6),ufo.iloc[[0,2,5],[0,2,4]])
```

```python
# selecting multiple coulmns in two different way
display(ufo[['City','Shape Reported']],ufo.loc[:,['City','Shape Reported']])
```

```python
# more confucing type interpretation
# it is kind of iloc in row and no rules coulmn wise. So better to avoid this kid of notation
ufo[0:2]
```

```python
# let's now learn the .ix 
# the .ix do the both the label that is .loc and integer postion that is .loc both.
# for both row and coulmns it will check is it label or integer postion. If label will do label wise, if postion 
# will do postion wise
drinks = pd.read_csv("http://bit.ly/drinksbycountry",index_col= 'country')
display(drinks.head(5),drinks.index.name)
```

```python
#drinks.reset_index(level = ['country'],inplace = True)
# drinks.head(5)
```

```python
display(drinks.ix['Albania',0],drinks.ix[1,'beer_servings'],drinks.ix['Albania':'Angola', 0:3])
```

```python
# drinks.ix[0:2,0:2] , here in the row and coulmn .iloc or integer postion so starting postion 0 is included and 
# 2 is excluded
display(drinks.head(6),drinks.ix[0:2,0:2])
```

```python
# here the  .ix is quite confusing as in row it is working as loc and in columns working as .iloc
# though we have given the input as integer for both the row and coulmnns
# this is because when for a string we use integer it uses postion and uses .iloc but if we are using 
# integer postions for numeric data then it is using Labale and uses .loc instead of .iloc
display(ufo.head(6),ufo.ix[0:2,0:2])
```
