```python
import os
os.chdir('/Users/rajatpa/Documents/DS/code_basics/py/pandas/8_concat')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
india_weather = pd.DataFrame({'city': ['mumabi','delhi','bangalore'] ,
                             'temperature': [32,45,30],
                             'humidity': [80,60,78]
                             })

india_weather                           
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather],axis = 'columns')
df
```

```python
df1 = pd.concat([india_weather,us_weather],axis = 1)
df1
```

```python
df2 = pd.concat([india_weather,us_weather],axis = 'rows')
df2
```

```python
# If we want continous index then we can use ignore_index = True
df2 = pd.concat([india_weather,us_weather],axis = 'rows',ignore_index = True)
df2
```

```python
# If we want continous index then we can use ignore_index = True
df3 = pd.concat([india_weather,us_weather],axis = 0,ignore_index = True)
df3
```

```python
# If we want to give key to the concatenated DataFrame while creating but we canot use the ignore_index = True
df4 = pd.concat([india_weather,us_weather],axis = 0,keys = ['india','usa'])
display(df4,df4.index.name ,df4.index.values)
```

```python
display(df4.loc[[('india', 0)],:],df4.loc[[('india', 1)],:],df4.loc[[('india', 2)],:])
```

```python
display(df4.loc[[('usa', 0)],:],df4.loc[[('usa', 1)],:],df4.loc[[('usa', 2)],:])
```

```python
df5 = pd.concat([india_weather,us_weather],axis = 1,keys = ['india','usa'])
df5
```

```python
display(df5.columns.names,df5.columns,df5.loc[:,"india"],df5.loc[:,"india"].loc[:,["city"]])
```

```python
display(df5.columns.names,df5.columns,df5.loc[:,"usa"].loc[:,['temperature']])
```

```python
display(df5['india'],df5.india)
```

```python
display(df5['usa'],df5.usa)
```

```python
# If we will use keys while creating concatenate DataFrame row wise then this is act as indexing
df6 = pd.concat([india_weather,us_weather],axis = 0,keys = ['india','usa'])
df6
```

```python
display(df6.loc['india'],df6.loc['usa'])
```

```python
# Incase of the cities order in the two DataFrame are different, then we will not get them in order
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
})
temperature_df
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
})
windspeed_df

display(temperature_df,windspeed_df)
```

```python
df7 = pd.concat([temperature_df,windspeed_df],axis = 1)
df7
```

```python
#If we have to rectify the error then we have to use the index for ordering
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df

display(temperature_df,windspeed_df)
```

```python

df8 = pd.concat([temperature_df,windspeed_df],axis = 1)
df8
```

```python
df9 = pd.concat([temperature_df,windspeed_df],axis = 0)
df9
```

```python
# If we want to append a series to a DataFrame
s = pd.Series(['Humid','Dry',"rain"],name = 'Event')
s
```

```python
df10 = pd.concat([temperature_df,s],axis = 1)
df10
```
