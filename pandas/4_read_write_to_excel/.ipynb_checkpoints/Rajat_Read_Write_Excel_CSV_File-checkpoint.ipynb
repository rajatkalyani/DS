```python
# If we have some inputs as not avaiable or na or something else, we have to convert that to NaN
df7 = pd.read_csv('/Users/rajatpa/Documents/DS/code_basics/py/pandas/4_read_write_to_excel/stock_data.csv',na_values = ['not available','n.a.'])
display(df,df7)
```

```python
# we can observe that there is a -1 values under revenue which is ideally not in real time scenario but we have 
# another -1 which is earn per share (eps) value which is fine.
# so if we replace -1 with NaN then both will replace which we donot want. 
# To resolve this issue we can use dictionary
# here basically we are telling if you find any of the values not available' or 'n.a.' or -1
# convert this to NaN for revenue
df8 = pd.read_csv('/Users/rajatpa/Documents/DS/code_basics/py/pandas/4_read_write_to_excel/stock_data.csv',
                  na_values ={'eps': ['not available','n.a.'],
                              'revenue' :['not available','n.a.',-1],
                             'people': ['not available','n.a.']})
                  
df8
```

```python
# Now let's do some operation with excel file
df9 = pd.read_excel('/Users/rajatpa/Documents/DS/code_basics/py/pandas/4_read_write_to_excel/stock_data.xlsx','Sheet1')
df9
```

```python
# now if we want to convert the n.a of people column to 'sam walton', then we have use converters argument 
def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
    
df10 = pd.read_excel('/Users/rajatpa/Documents/DS/code_basics/py/pandas/4_read_write_to_excel/stock_data.xlsx','Sheet1'
                    ,converters = { 'people': convert_people_cell,
                                     'eps' :convert_eps_cell })
df10    
```

```python
# writing to a excel file
# startrow = 1, startcol = 2 means, it will leave one row and leave two column
#df10.to_excel("stock_data_modified.xlsx", sheet_name = 'stocks',index = False,startrow = 1, startcol = 2 )
df10.to_excel("stock_data_modified.xlsx", sheet_name = 'stocks',index = False,startrow = 1, startcol = 2 )
```

```python
# Write two dataframes to two separate sheets in excel
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
