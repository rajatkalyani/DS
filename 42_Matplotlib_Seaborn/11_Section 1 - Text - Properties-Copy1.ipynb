# Text: Properties

```python
# So, let's import required packages
import matplotlib.pyplot as plt

# We need Text artist
from matplotlib.text import Text

%matplotlib inline
```

```python
txt1 = Text(text='Over here', x=0.5, y=0.5)
txt2 = Text(text='Another Dude', x=0.2, y=0.2)

fig, ax = plt.subplots()

t1 = ax._add_text(txt1)
t2 = ax._add_text(txt2)
```

Take your time and read through the documentation.
Assuming you have done so, let's  fine tune text properties

```python
t1.set_color('red') # Sets font color
t1.set_size(24) # set font size
t1.set_alpha(0.8) # set opacity
t1.set_family('sans-serif') # set font family
t1.set_linespacing(0.2) 
t1.set_rotation(45)
t1.set_style('italic')
t1.set_weight('bold')
t1.set_x(0.4)
t1.set_y(0.8)


t2.set_position((0.9, 0.5))

ax.figure
```

Time to customize the wrapping box.

```python
t1.set_bbox(dict(facecolor='black', 
                 boxstyle='darrow, pad=0.4', alpha=0.7, hatch='o' ))
t1.figure
```
