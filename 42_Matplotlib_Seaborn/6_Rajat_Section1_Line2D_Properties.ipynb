```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
np.random.seed(5)
x = np.linspace(1, 40, 10)

y1 = x / (np.random.rand(10)*2)
y2 = x / (np.random.rand(10)*3)


line1 = Line2D(x, y1, label='Son of a Line')
line2 = Line2D(x, y2, label='Another Line')

fig, ax = plt.subplots()


ax.add_line(line1)
ax.add_line(line2)

ax.autoscale()
```

```python
# how to increase the figure size
# We can use help function to find out the method and attributes. 
help(fig)
```

```python
fig.set_size_inches(15,5)
fig
```

```python
help(line1)
```

```python
# Let's create handlers for the two lines
# in the axess, as we will be referring to them
# soon
l1 = ax.lines[0]
l2 = ax.lines[1]
```

```python
# how about we change line1 opacity?
# The alpha method decides the transperancy of the line
l1.set_alpha(0.3)
fig
```

```python
for line in ax.lines:
    # make all the lines red
    line.set_color('red')
    ## make all the lines wider
    line.set_lw(2)
fig
```

```python
# Let's make our first line
# dashed
l1.set_linestyle('dashed')
fig
```

```python
# let's change the line2 style to star
l2.set_marker('*')
fig
```

```python
# Now let's modify various properties of the marker
l2.set_markersize(20)
l2.set_markeredgecolor('red')
l2.set_markerfacecolor('white')
l2.set_markeredgewidth(1)
# here we are metioning that every defined point in the x-axis have marker(0,5,10 ... 40)
# if we write l2.set_markevery(2), then markers will at point 0,10,20,30 and 40
l2.set_markevery(1)
# just to increase the look and feel
l2.set_sketch_params(0.9)
fig
```

```python
# We want the line1 which is the dashed line should go beyond the right border 
# for doing this we have to make the cliping as False
# Also we have to decrease the x-axis boundry
l1.set_clip_on(False)
# here we are saying, turn off the clipping from the dashed line
ax.set_xlim(0, 39)
# set xaxis boundary to 39
fig
```

```python
ax.set_title("Some line chart")
ax.set_xlabel('This is x axis')
ax.set_ylabel('This is y axis')

ax.figure
```
